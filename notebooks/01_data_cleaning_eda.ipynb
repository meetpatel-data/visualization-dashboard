```python
import pandas as pd

df = pd.read_csv("../data/raw/SuperStore_Sales_Dataset.csv")
df.head()
```

```python
df.info()
df.isna().sum().sort_values(ascending=False).head(15)
df.describe(include="all").T
```

```python
df[["Order Date", "Ship Date"]].head(10)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True, errors="coerce")
df["Ship Date"]  = pd.to_datetime(df["Ship Date"],  dayfirst=True, errors="coerce")
```

```python
df[["Order Date", "Ship Date"]].head()
```

```python
df["order_year"] = df["Order Date"].dt.year
df["order_month"] = df["Order Date"].dt.to_period("M").astype(str)
df["ship_days"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
monthly = df.groupby("order_month", as_index=False)[["Sales", "Profit"]].sum()
top_products = (df.groupby("Product Name", as_index=False)[["Sales", "Profit"]].sum()
                  .sort_values("Sales", ascending=False).head(10))
top_regions = (df.groupby("Region", as_index=False)[["Sales", "Profit"]].sum()
                 .sort_values("Sales", ascending=False))

monthly.head(), top_products, top_regions
```

```python
df.to_csv("../data/processed/superstore_clean.csv", index=False)
```
